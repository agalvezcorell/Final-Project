```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import re
```

Web scrapping de la página de pantone para sacar los colores y su código hexadecimal

```python
url = "https://www.logorapid.com/pantone"
```

```python
datos = requests.get(f"{url}").text
soup = BeautifulSoup(datos, 'html.parser')
extraigo = str(soup.select('div.colortable'))
```

```python
divido = extraigo.split("<p>")
todos = divido[1::2]
```

```python
def matches(string):
    nombre = re.findall(r"^\w+\s*\w*",string)
    codigo = re.findall(r"(\#\w*)",string)
    return nombre[0],codigo[0]
```

```python
columnas = ["color","codigo"]
colores = pd.DataFrame(list(map(matches,todos)),columns = columnas)
```

```python
colores
```

```python
def hex_to_rgb(value):
    value = value.lstrip('#')
    lv = len(value)
    return tuple(int(value[i:i + lv // 3], 16) for i in range(0, lv, lv // 3))
```

```python
colores["RGB"] = colores.codigo.map(hex_to_rgb)
```

```python
def extraeR(x):
    b = x[0]
    return b 
def extraeG(x):
    b = x[1]
    return b 
def extraeB(x):
    b = x[2]
    return b 
colores["R"] = colores.RGB.apply(extraeR)
colores["G"] = colores.RGB.apply(extraeG)
colores["B"] = colores.RGB.apply(extraeB)
```

```python
colores["visual"] = ""
```

```python
styled = colores.style.apply(lambda col: ['background-color: %s' % (colores.iloc[i].codigo) for i in range(col.size)], subset=['visual'])
```

```python
colores.to_csv("data/colores.csv")
```
